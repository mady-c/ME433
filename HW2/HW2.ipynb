```python
[t,s] = getdata("D")
whichplot = "D"

#fast fourier transform with setup
def FFT(t,s):
    Fs = len(t)/t[-1] # sample rate
    Ts = 1.0/Fs; # sampling interval
    ts = np.arange(0,t[-1],Ts) # time vector
    y = s # the data to make the fft from
    n = len(y) # length of the signal
    k = np.arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    frq = frq[range(int(n/2))] # one side frequency range
    Y = np.fft.fft(y)/n # fft computing and normalization
    Y = Y[range(int(n/2))]
    return([Y,frq,n])

#fast fourier transform with setup already done (sfiltered is filtered signal)
def filteredFFT(n,sfiltered):
    Yfiltered = np.fft.fft(sfiltered)/n # fft computing and normalization
    Yfiltered = Yfiltered[range(int(n/2))]
    return(Yfiltered)

y = s
[Y,frq,n] = FFT(t,s)


fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(t,y,'k')
fig.suptitle('Signal ' + whichplot + ' vs Time')
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax2.loglog(frq,abs(Y),'k') # plotting the fft
ax2.set_xlabel('Freq (Hz)')
ax2.set_ylabel('|Y(freq)|')
fig.set_size_inches(10,10)
# plt.savefig('Signal'+whichplot+'_fft.png') #un-comment when you want to save file
plt.show()
```

```python
# moving average filter

t,s = getdata("C")
whichplot = "C"
X = 50

sMAF = np.zeros(len(s))

tempstor = np.zeros(X); #moving vector of saved points

for ii in range(len(s)):
    tempstor[0:-1] = tempstor[1:]  #moves previous data back a slot
    tempstor[-1] = s[ii] #last slot is newest point
#     sMAF[ii] = sum(tempstor)/len(tempstor)
    MAFval = sum(tempstor)/len(tempstor)
    sMAF[ii] = MAFval
    
[Y,frq, n] = FFT(t,s) 
y = s
yMAF = sMAF
YMAF = filteredFFT(n,sMAF) 

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(t,y,'k', label='raw data')
ax1.plot(t,yMAF, 'r', label = 'filtered data')
ax1.legend()
fig.suptitle('Signal ' + whichplot + ' vs Time')
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax1.set_title('MAF with ' + str(X) + ' data points being averaged')
ax2.loglog(frq,abs(Y),'k', label='raw data') # plotting the fft
ax2.loglog(frq,abs(YMAF),'r', label = 'filtered data')
ax2.legend()
ax2.set_xlabel('Freq (Hz)')
ax2.set_ylabel('|Y(freq)|')
fig.set_size_inches(10,10)
plt.savefig('Signal'+whichplot+'_MAF_fft.png')
plt.show()

## SAVE FIGURES WITH LEGEND AND # of DATA POINTS
```

Now try low-pass filtering with an IIR. Write a loop that adds a value into the average with the two weights A and B, so that new_average[i] = A * new_average[i-1] + B * signal[i]. Note that A+B=1, otherwise the signal will get bigger with time (A+B>1) or go to 0 (A+B<1).
```python
# infinite impulse response (IIR)
t,s = getdata("A") 
whichplot = "A"
y = s
A = .998
B = 1 - A 

new_average = np.zeros(len(s))
for ii in range(1,len(s)):
    new_average[ii] = A*new_average[ii-1] + B*s[ii]
    
[Y,frq, n] = FFT(t,s) 
YIIR = filteredFFT(n,new_average)

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(t,y,'k', label='raw data')
ax1.plot(t,new_average, 'r', label = 'filtered data')
ax1.legend()
fig.suptitle('Signal ' + whichplot + ' vs Time')
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax1.set_title('IIR with A = ' + str("{:.3f}".format(A))+ ' and B = ' + str("{:.3f}".format(B))) ## had weird issues w floating point display
ax2.loglog(frq,abs(Y),'k', label='raw data') # plotting the fft
ax2.loglog(frq,abs(YIIR),'r', label = 'filtered data')
ax2.legend()
ax2.set_xlabel('Freq (Hz)')
ax2.set_ylabel('|Y(freq)|')
fig.set_size_inches(10,10)
plt.savefig('Signal'+whichplot+'_IIR.png')
plt.show()
```
